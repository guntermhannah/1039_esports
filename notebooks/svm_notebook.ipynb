# SVM Model Testing Notebook

```python
import pandas as pd
import numpy as np
import os

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import RobustScaler
from sklearn.svm import SVR, SVC
```

```python
df = pd.read_csv(os.path.join('data', 'player_pairs_stats.csv'))
df.drop_duplicates(subset=['match_id'], inplace=True)
df
```

```python
for row in range(len(df)):
    if df['player'][row] == df['winner'][row]:
        df['winner'][row] = 1
    else:
        df['winner'][row] = 0
df
```

```python
X = df.drop(columns=['match_id', 'player', 'player_hero_id', 'opponent', 'opponent_hero_id', 'winner'])
y = df['winner']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
scaler = RobustScaler()
scaler.fit(X_train, y_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(len(X_train_scaled))
print(len(y_train))
```

```python
svr = SVR()
svr.fit(X_train_scaled, y_train)
```

```python
results = cross_val_score(svr, X_train_scaled, y_train).mean()
results
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
```

```python
results_svc = cross_val_score(svr, X_train_scaled, y_train).mean()
results_svc
```

```python
test = cross_val_score(svr, X_test_scaled, y_test).mean()
test
```

```python
from sklearn.model_selection import GridSearchCV
svr = SVR()

search = {
    'kernel':['sigmoid', 'rbf'],
    'C':[0.01, 0.1, 1, 10],
    'gamma':[0.1, 1, 10]
}

rsearch = GridSearchCV(svr, search, n_jobs=-1, scoring='accuracy', cv=3, verbose=0)
rsearch.fit(X_train_scaled, y_train)
```
